```python
class ZomatoApi():

    def __init__(self,API_KEY):
        self.API_KEY = API_KEY
        self.base_url = "https://developers.zomato.com/api/v2.1/"
        # the api key is passed in the header of the url
        self.headers = {"user-key":"{}".format(self.API_KEY)}

    def search(self,res_id):
        self.res_id = res_id
        self.search_url = self.base_url + "search"

        self.params = {
            'res_id' : "{}".format(self.res_id),
            }

        self.response = requests.get(self.search_url,headers=self.headers,params=self.params)
        return self.response.json()

    def getRestaurantDetails(self,jsonObj):
        final_list = []
        for n in range(0,len(jsonObj['restaurants'])-1):
            rid = jsonObj["restaurants"][n]['restaurant']['R']['res_id']
            rName = jsonObj["restaurants"][n]['restaurant']['name']
            rImg = jsonObj["restaurants"][n]['restaurant']['featured_image']
            rRate = jsonObj["restaurants"][n]['restaurant']['user_rating']['aggregate_rating']
            rRateType = jsonObj["restaurants"][n]['restaurant']['user_rating']['rating_text']
            rVotes = jsonObj["restaurants"][n]['restaurant']['user_rating']['votes']            
            rreviews = jsonObj["restaurants"][n]['restaurant']['all_reviews']['reviews']
            rreview_count = jsonObj["restaurants"][n]['restaurant']['all_reviews_count']
            rLocation = jsonObj["restaurants"][n]['restaurant']['location']['locality']
            rcity = jsonObj["restaurants"][n]['restaurant']['location']['city']
            rcityid = jsonObj["restaurants"][n]['restaurant']['location']['city_id']
            rLat = jsonObj["restaurants"][n]['restaurant']['location']['latitude']
            rLong = jsonObj["restaurants"][n]['restaurant']['location']['longitude']
            rPrice = jsonObj["restaurants"][n]['restaurant']['price_range']
            rCurrency = jsonObj["restaurants"][n]['restaurant']['currency']
            ravgcost = jsonObj["restaurants"][n]['restaurant']['average_cost_for_two']
            rcuisine = jsonObj["restaurants"][n]['restaurant']['cuisines']
            rphotos_url = jsonObj["restaurants"][n]['restaurant']['photos_url']
            rphotos_count = jsonObj["restaurants"][n]['restaurant']['photo_count']
            rphone = jsonObj["restaurants"][n]['restaurant']['phone_numbers']
            restablishment = jsonObj["restaurants"][n]['restaurant']['establishment']
            rOnline = jsonObj["restaurants"][n]['restaurant']['has_online_delivery']
            rBook = jsonObj["restaurants"][n]['restaurant']['has_table_booking']
            final_list.append([rid,rName,rImg,rRate,rRateType,rVotes,rreview_count,rLocation,rcity,rcityid,rLat,
                               rLong,rPrice,rCurrency,ravgcost,rcuisine,rphotos_url,rphotos_count,rphone,restablishment,
                              rOnline,rBook])
        return final_list
```
