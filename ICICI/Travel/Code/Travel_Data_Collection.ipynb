```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
import time
import re
import sys
import time
```

```python
class travel_Api():

    def __init__(self,rapidapi_key,rapidapi_host):
        self.rapidapi_key = rapidapi_key
        self.rapidapi_host = rapidapi_host
        self.base_url1 = "https://skyscanner-skyscanner-flight-search-v1.p.rapidapi.com/apiservices/reference/v1.0/countries/en-US"
        self.base_url2 = "https://skyscanner-skyscanner-flight-search-v1.p.rapidapi.com/apiservices/browsequotes/v1.0/IN/USD/en-US/DEL-sky/"
        self.base_url3 = "https://www.travel-advisory.info/api"
        self.base_url4 = "https://travel-hacking-tool.p.rapidapi.com/api/airports/"
        self.base_url5 = "https://iatacodes-iatacodes-v1.p.rapidapi.com/api/v5/airports"
        self.base_url6 = "https://iatacodes-iatacodes-v1.p.rapidapi.com/api/v5/cities"
        self.base_url7 = "https://currencyscoop.p.rapidapi.com/latest"
        
        
        
        # the api key is passed in the header of the url
        self.headers = {"x-rapidapi-key":"{}".format(self.rapidapi_key),
                       "x-rapidapi-host":"{}".format(self.rapidapi_host)}
          
    # The List of Country Codes and Countries
    
    def list_countries(self):
        
        self.search_url1 = self.base_url1

        self.response = requests.get(self.search_url1,headers=self.headers)
        print(self.response.url)
        return self.response.json()
    
    def getcountryList(self,jsonObj):
        
        if 'code' in jsonObj.keys():
            
            return 'Invalid Request'
        
        else:        
            
            cols = ['Country_Code','Country']
            
            Countries_List = pd.DataFrame(columns = cols)
            
            for i in range(len(jsonObj['Countries'])):
                
                list = []
                
                Code = jsonObj['Countries'][i]['Code']
                Name = jsonObj['Countries'][i]['Name']
                
                list.append([Code,Name])

                for i in range(len(list)):    
                    df_length = len(Countries_List)
                    Countries_List.loc[df_length] = list[i]
                
            return Countries_List
        
    # The List of Country Codes, Countries, Travel Advisory
    
    def Travel_Advisory(self):
        
        self.search_url = self.base_url3

        self.response = requests.get(self.search_url,headers=self.headers)
        print(self.response.url)
        return self.response.json()
    
    def getTravelAdvisorydetails(self,jsonObj):
        
        if 'code' in jsonObj.keys():
            
            return 'Invalid Request'
        
        else:        
            
            cols = ['Country_Code','Country','Continent','Advisory_Score','Advisory_Message']
            
            Country_Advisory_List = pd.DataFrame(columns = cols)
            
            for key in jsonObj['data'].keys():
                
                list = []
                
                Code = jsonObj['data'][key]['iso_alpha2']
                Name = jsonObj['data'][key]['name']
                Continent = jsonObj['data'][key]['continent']
                Advisory_Score = jsonObj['data'][key]['advisory']['score']
                Advisory_Message = jsonObj['data'][key]['advisory']['message']
                
                list.append([Code,Name,Continent,Advisory_Score,Advisory_Message])

                for i in range(len(list)):    
                    df_length = len(Country_Advisory_List)
                    Country_Advisory_List.loc[df_length] = list[i]
                
            return Country_Advisory_List
        
    # The List of Airports, City Codes, Country Codes
    
    def AirportDetails(self):
        
        self.search_url = self.base_url4

        self.response = requests.get(self.search_url,headers=self.headers)
        print(self.response.url)
        return self.response.json()
    
    def getAirportDetails(self,jsonObj):
        
        if 'code' in jsonObj:
            
            return 'Invalid Request'
        
        else:        
            
            cols = ['Airport_Code','Airport_Name','Latitude','Longitude','Time_Zone','City_Code','Country_Code']
            
            Airport_Details = pd.DataFrame(columns = cols)
            
            for i in range(len(jsonObj)):
                
                list = []
                
                Airport_Code = jsonObj[i]['code']
                Airport_Name = jsonObj[i]['name']
                Latitude = jsonObj[i]['latitude']
                Longitude = jsonObj[i]['longitude']
                Time_Zone = jsonObj[i]['time_zone']
                City_Code = jsonObj[i]['city_code']
                Country_Code = jsonObj[i]['country']
                
                list.append([Airport_Code,Airport_Name,Latitude,Longitude,Time_Zone,City_Code,Country_Code])

                for i in range(len(list)):    
                    df_length = len(Airport_Details)
                    Airport_Details.loc[df_length] = list[i]
                
            return Airport_Details
    
    
    def AirportList(self,api_key):
        self.api_key = api_key
        
        self.search_url = self.base_url5
        
        self.params = {
            'api_key' : "{}".format(self.api_key)            
            }

        self.response = requests.get(self.search_url,headers=self.headers,params=self.params)
        print(self.response.url)
        return self.response.json()
    
    def getAirportList(self,jsonObj):
        
        if 'code' in jsonObj.keys():
            
            return 'Invalid Request'
        
        else:        
            
            cols = ['Airport_Code','Airport_Name']
            
            Airport_List = pd.DataFrame(columns = cols)
            
            for i in range(len(jsonObj['response'])):
                
                list = []
                
                Airport_Code = jsonObj['response'][i]['code']
                Airport_Name = jsonObj['response'][i]['name']
                
                list.append([Airport_Code,Airport_Name])

                for i in range(len(list)):    
                    df_length = len(Airport_List)
                    Airport_List.loc[df_length] = list[i]
                
            return Airport_List
    
    # Get the list of Cities
    
    def CityList(self,api_key):
        self.api_key = api_key
        
        self.search_url1 = self.base_url6
        
        self.params = {
            'api_key' : "{}".format(self.api_key)            
            }

        self.response = requests.get(self.search_url1,headers=self.headers,params=self.params)
        print(self.response.url)
        return self.response.json()
    
    def getCityList(self,jsonObj):
        
        if 'code' in jsonObj.keys():
            
            return 'Invalid Request'
        
        else:        
            
            cols = ['City_Code','Country_Code','City_Name']
            
            City_List = pd.DataFrame(columns = cols)
            
            for i in range(len(jsonObj['response'])):
                
                list = []
                
                City_Code = jsonObj['response'][i]['code']
                City_Name = jsonObj['response'][i]['name']
                Country_Code = jsonObj['response'][i]['country_code']
                
                list.append([City_Code,Country_Code,City_Name])

                for i in range(len(list)):    
                    df_length = len(City_List)
                    City_List.loc[df_length] = list[i]
                
            return City_List
    
    # Get Flight Prices
    def browse_flightquotes(self,location,outbound_date,inbound_date):
        self.location = location
        self.outbound_date = outbound_date
        self.inbound_date = inbound_date
        self.search_url = self.base_url2 + location + '/' + outbound_date
        
        self.params = {
            'inboundpartialdate' : "{}".format(self.inbound_date)
            }

        self.response = requests.get(self.search_url,headers=self.headers,params=self.params)
        #print(self.response.url)
        return self.response.json()
    
    def getflightpricedetails(self,jsonObj):
        
        #print(jsonObj)
        if 'errors' in jsonObj.keys():
            
            return 'Invalid Request'
        
        elif 'message' in jsonObj.keys():
            
            return 'Limit Exceeded'         
            
        else:        
            
            cols = ['Price']
            
            Price_List = pd.DataFrame(columns = cols)
            
            for i in range(len(jsonObj['Quotes'])):
                
                list = []
                
                Price = jsonObj['Quotes'][i]['MinPrice']
                
                list.append([Price])

                for i in range(len(list)):    
                    df_length = len(Price_List)
                    Price_List.loc[df_length] = list[i]
                
            return Price_List
    
    #Currency Converter
    def CurrencyConverter(self):
        
        self.search_url = self.base_url7
        
        self.response = requests.get(self.search_url,headers=self.headers)
        print(self.response.url)
        
        return self.response.json()
    
    def getCurrencydetails(self,jsonObj):
        
        if 'code' in jsonObj.keys():
            
            return 'Invalid Request'
        
        else:        
            
            Currency_List = pd.DataFrame()
            
            #print((jsonObj['response']['rates'].keys()))
            #print((jsonObj['response']['rates'].values()))
            Currency_List['Currency'] = (jsonObj['response']['rates'].keys())
            Currency_List['RatiotoUSD'] = (jsonObj['response']['rates'].values())
        
            return Currency_List
```

```python
from re import sub
from gensim.utils import simple_preprocess
import gensim.downloader as api
from gensim.corpora import Dictionary
from gensim.models import TfidfModel
from gensim.models import WordEmbeddingSimilarityIndex
from gensim.similarities import SparseTermSimilarityMatrix
from gensim.similarities import SoftCosineSimilarity
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from nltk import word_tokenize          
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import stopwords
import pandas as pd
```

```python
# Load the model: this is a big file, can take a while to download and open
glove = api.load("glove-wiki-gigaword-50")    
similarity_index = WordEmbeddingSimilarityIndex(glove)
```

```python
# Download stopwords list
stopwords = nltk.corpus.stopwords.words('english')
stopwords.append('Known')
stopwords.append('Recommended')
stopwords.append('places')
stopwords.append('visit')
stopwords.append('recommended')
stopwords.append('known')
```

```python
# From: https://github.com/RaRe-Technologies/gensim/blob/develop/docs/notebooks/soft_cosine_tutorial.ipynb
def preprocess(doc):
    # Tokenize, clean up input document string
    doc = sub(r'<img[^<>]+(>|$)', " image_token ", doc)
    doc = sub(r'<[^<>]+(>|$)', " ", doc)
    doc = sub(r'\[img_assist[^]]*?\]', " ", doc)
    doc = sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', " url_token ", doc)
    return [token for token in simple_preprocess(doc, min_len=0, max_len=float("inf")) if token not in stopwords]
```
