```python
class TriposoApi():

    def __init__(self,ACCOUNT_ID,API_KEY):
        self.API_KEY = API_KEY
        self.ACCOUNT_ID = ACCOUNT_ID
        self.base_url = "https://www.triposo.com/api/20200803/"
        # the api key is passed in the header of the url
        self.headers = {"X-Triposo-Account":"{}".format(self.ACCOUNT_ID),
                       "X-Triposo-Token":"{}".format(self.API_KEY)}
        
    def search_location(self,id,fields):
        self.id = id
        self.fields = fields
        self.search_url = self.base_url + "location.json"

        self.params = {
            'id' : "{}".format(self.id),
            'fields' : "{}".format(self.fields)
            }

        self.response = requests.get(self.search_url,headers=self.headers,params=self.params)
        print(self.response.url)
        return self.response.json()
    
    def getlocationdetails(self,jsonObj):
        
        if 'code' in jsonObj.keys():
            
            return 'Location Not Found/Invalid Request'
        
        elif jsonObj['results'] == []:
            
            return 'Location Not Found/Invalid Request'
        
        else:        
            final_list = []
            Location_Det = jsonObj
            #print(Location_Det)
            Location_ID = Location_Det['results'][0]['id']
            Parent_ID = Location_Det['results'][0]['parent_id']
            Latitude = Location_Det['results'][0]['coordinates']['latitude']
            Longitude = Location_Det['results'][0]['coordinates']['longitude']
            Score = Location_Det['results'][0]['score']
            Country_ID = Location_Det['results'][0]['country_id']
            Type = Location_Det['results'][0]['type']
            Name = Location_Det['results'][0]['name']
            Snippet_Language = Location_Det['results'][0]['snippet_language_info']
            Snippet = Location_Det['results'][0]['snippet']
            
            if Location_Det['results'][0]['climate'] != None:
                
                Temp_Avg_Max_Jan = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][0]
                Temp_Avg_Max_Feb = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][1]
                Temp_Avg_Max_Mar = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][2]
                Temp_Avg_Max_Apr = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][3]
                Temp_Avg_Max_May = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][4]
                Temp_Avg_Max_Jun = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][5]
                Temp_Avg_Max_Jul = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][6]
                Temp_Avg_Max_Aug = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][7]
                Temp_Avg_Max_Sep = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][8]
                Temp_Avg_Max_Oct = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][9]
                Temp_Avg_Max_Nov = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][10]
                Temp_Avg_Max_Dec = Location_Det['results'][0]['climate']['temperature']['average_max']['months'][11]
                Temp_Avg_Min_Jan = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][0]
                Temp_Avg_Min_Feb = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][1]
                Temp_Avg_Min_Mar = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][2]
                Temp_Avg_Min_Apr = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][3]
                Temp_Avg_Min_May = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][4]
                Temp_Avg_Min_Jun = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][5]
                Temp_Avg_Min_Jul = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][6]
                Temp_Avg_Min_Aug = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][7]
                Temp_Avg_Min_Sep = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][8]
                Temp_Avg_Min_Oct = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][9]
                Temp_Avg_Min_Nov = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][10]
                Temp_Avg_Min_Dec = Location_Det['results'][0]['climate']['temperature']['average_min']['months'][11]
                
            else:
                Temp_Avg_Max_Jan = 'NA'
                Temp_Avg_Max_Feb = 'NA'
                Temp_Avg_Max_Mar = 'NA'
                Temp_Avg_Max_Apr = 'NA'
                Temp_Avg_Max_May = 'NA'
                Temp_Avg_Max_Jun = 'NA'
                Temp_Avg_Max_Jul = 'NA'
                Temp_Avg_Max_Aug = 'NA'
                Temp_Avg_Max_Sep = 'NA'
                Temp_Avg_Max_Oct = 'NA'
                Temp_Avg_Max_Nov = 'NA'
                Temp_Avg_Max_Dec = 'NA'
                Temp_Avg_Min_Jan = 'NA'
                Temp_Avg_Min_Feb = 'NA'
                Temp_Avg_Min_Mar = 'NA'
                Temp_Avg_Min_Apr = 'NA'
                Temp_Avg_Min_May = 'NA'
                Temp_Avg_Min_Jun = 'NA'
                Temp_Avg_Min_Jul = 'NA'
                Temp_Avg_Min_Aug = 'NA'
                Temp_Avg_Min_Sep = 'NA'
                Temp_Avg_Min_Oct = 'NA'
                Temp_Avg_Min_Nov = 'NA'
                Temp_Avg_Min_Dec = 'NA'

            final_list.append([Location_ID,Parent_ID,Name,Country_ID,Type,Snippet,Snippet_Language,
                               Temp_Avg_Min_Jan,Temp_Avg_Max_Jan,Temp_Avg_Min_Feb,Temp_Avg_Max_Feb,Temp_Avg_Min_Mar,
                               Temp_Avg_Max_Mar,Temp_Avg_Min_Apr,Temp_Avg_Max_Apr,Temp_Avg_Min_May,Temp_Avg_Max_May,
                               Temp_Avg_Min_Jun,Temp_Avg_Max_Jun,Temp_Avg_Min_Jul,Temp_Avg_Max_Jul,Temp_Avg_Min_Aug,
                               Temp_Avg_Max_Aug,Temp_Avg_Min_Sep,Temp_Avg_Max_Sep,Temp_Avg_Min_Oct,Temp_Avg_Max_Oct,
                               Temp_Avg_Min_Nov,Temp_Avg_Max_Nov,Temp_Avg_Min_Dec,Temp_Avg_Max_Dec])
            cols = ['Location_ID','Parent_ID','Name','Country_ID','Type','Snippet','Snippet_Language',
                    'Temp_Avg_Min_Jan','Temp_Avg_Max_Jan','Temp_Avg_Min_Feb','Temp_Avg_Max_Feb','Temp_Avg_Min_Mar',
                    'Temp_Avg_Max_Mar','Temp_Avg_Min_Apr','Temp_Avg_Max_Apr','Temp_Avg_Min_May','Temp_Avg_Max_May',
                    'Temp_Avg_Min_Jun','Temp_Avg_Max_Jun','Temp_Avg_Min_Jul','Temp_Avg_Max_Jul','Temp_Avg_Min_Aug',
                    'Temp_Avg_Max_Aug','Temp_Avg_Min_Sep','Temp_Avg_Max_Sep','Temp_Avg_Min_Oct','Temp_Avg_Max_Oct',
                    'Temp_Avg_Min_Nov','Temp_Avg_Max_Nov','Temp_Avg_Min_Dec','Temp_Avg_Max_Dec']

            loc = pd.DataFrame(columns = cols)
            for i in range(len(final_list)):    
                df_length = len(loc)
                loc.loc[df_length] = final_list[i]

            return loc        
```
